## Step 2: Cleaning the data set

Based on the above findings, we intend to do the following data cleaning:

a) drop column view: as it contains mostly zeros which does not make intuitive sense

b) drop column sqrt_basement: as it contains invalid data '?'. Alternative choice here is to turn the ? into Nulls and convert the data type to integer. However, given that this only contains part of the data about the size of the house and we already have the more complete column sqft_living available, we would not need to use sqft_basement. If we cleaned sqft_basement and it turned out to be highly correlated with house prices, it would also be correlated with sqft_living, and we'd prefer to use sqft_living in the model as it makes more intuitive sense as a predictor.

c) convert date from string object to datetime

d) create a new binary data series based on yr_renovated, similar to waterfront. Intuitively it would make sense that recently renovated houses sell for more.

Later on, before building the model, we will delete the rows that contain Nulls in waterfront.

```python
house_df['is_renovated'] = house_df['yr_renovated'].map(
                                                    lambda x: 0 if not x else (0 if math.isnan(x) else 1))

clean_df = house_df.drop(columns = ['view','sqft_basement'])

clean_df.date = clean_df.date.map(lambda x: pd.to_datetime(x))
```

#### Question 1: Have houses gotten bigger over the years?

This question is best explored by plotting graphs. On the x-axis we will plot yr_built and on the y-axis we have several data series representing the size of the property. We feel that sqft_living and bedrooms are the most representitive of the house size.

We first group data by yr_built and plot the median sqft_living per each yr_built, this reduces/summerises the data which would enable an easier understanding on any trends present. We choose median for sqft_living, as this is a continous data set with some extremely large values that can skew the mean. The bedrooms column contains categorical data. The median value per year can easily be the same for different values of year. Therefore we chose to plot the mean number of bedrooms against the year the properties were built, in order to spot any trends.

We will also look at number of bathrooms as a supplement to the size trend.

```python

houses_by_year = clean_df.groupby('yr_built')
houses_in_year = houses_by_year.size()
```

```python
avg_live_size_by_year = clean_df.groupby('yr_built')['sqft_living'].median()
avg_room_size_by_year = clean_df.groupby('yr_built')['bedrooms'].mean()
avg_bath_size_by_year = clean_df.groupby('yr_built')['bathrooms'].mean()
```

```python
plt.style.use('bmh')
fig = plt.figure(figsize = (18,12))

ax1 = fig.add_subplot(221)
plt.scatter(avg_live_size_by_year.index, avg_live_size_by_year.values)
ax1.set_title('Median Living Size by Year Built')
ax1.set_xlabel('Year built')
ax1.set_ylabel('Living Space (SqFt)')

ax2 = fig.add_subplot(222)
plt.scatter(avg_room_size_by_year.index, avg_room_size_by_year.values)
ax2.set_title('Average No. of Bedrooms by Year Built')
ax2.set_xlabel('Year built')
ax2.set_ylabel('Number of Bedrooms')

ax3 = fig.add_subplot(223)
plt.scatter(avg_bath_size_by_year.index, avg_bath_size_by_year.values)
ax3.set_title('Average No. of Bathrooms by Year Built')
ax3.set_xlabel('Year built')
ax3.set_ylabel('Number of Bathrooms')

ax4 = fig.add_subplot(224)
plt.bar(houses_in_year.index, houses_in_year.values)
ax4.set_title('Properties Built per Year')
ax4.set_xlabel('Year built')
ax4.set_ylabel('Properties Built')
plt.show()
```

We next look to explore the relationship between the independent variables and dependent variable.

```python
# create a function to plot scatter plots between Y and independent variables
# note we plot graphs for all data columns but certain columns we dont plan to use as predictors 
# like id, date (not valid), lat, long, zipcode

def plot_data(dataframe, y):
    fig = plt.figure(figsize = (10,100))
    df_less_y = dataframe.drop(y, axis = 1)
    n = len(df_less_y.columns)
    cols = list(zip(range(n),df_less_y.columns))
    for i in range(n):
        ax = fig.add_subplot(round(n/2,0), 2, i+1)
        ax.scatter(df_less_y[cols[i][1]],dataframe[y], marker='.', alpha=0.1)
        ax.set_title('Plot of {} against {}'.format(y,cols[i][1]))
        ax.set_xlabel(cols[i][1])
        ax.set_ylabel(y)
```

```python
plot_data(clean_df.drop('date', axis = 1), 'price')
```

From the plots of x variables with price, looks like sqft_living or sqft_living15, bathrooms and grade has the strongest relationship with price. In the case of the discrete variable grade, as the grade icreases, the whole range of price points moves upwards.

Model explains about 65% of the spread of price, in the data we excluded 44 properties that are over 3 million as we think the fundamentals of such properties are very different.
```python
# define a function to store the equation
def predict_price(sqft_living, waterfront, lat):
    y_log = -70.1387 + 0.8026 * np.log(sqft_living) + 0.8021 * waterfront + 1.6216 * lat
    price1 = np.exp(y_log)
    return price1

```

## Step 7: Further Insights into Zipcodes and Areas

Given that our model's relatively low prediction power (64% R2), it can only be used as a rough guideline to predict the price.

One important factor that people care about when considering a house is the location, chosen based on factors like affordability, 'feel' of the neighbourhood, ease of transport, schools etc. We intend to give more insights on the first two in the below analysis. Again, we are assuming our dataset is representive of the general housing stock existing.


```python

zipcode_info_dict = {}
med_prices = []
indices = []
num_of_expensive_houses_list = []
num_of_houses_list = []
longitudes = []
latitudes = []
percentage_of_expensive_houses = []


for code in clean_df.zipcode.unique():
    indices.append(code)
    med_prices.append(clean_df.groupby('zipcode').median().loc[code,'price'])
    num_of_expensive_houses = len(clean_df[(clean_df['zipcode'] == code) & (clean_df['price'] >= 1000000)])
    num_of_expensive_houses_list.append(num_of_expensive_houses)
    num_of_houses = len(clean_df[clean_df['zipcode'] == code])
    num_of_houses_list.append(num_of_houses)
    longitudes.append(
                    clean_df.groupby('zipcode').median().loc[code,'long'])
    latitudes.append(
                    clean_df.groupby('zipcode').median().loc[code,'lat'])
    percentage_of_expensive_houses.append(
                                        round((num_of_expensive_houses/num_of_houses*100),2))
    zipcode_info_dict['med_prices'] = med_prices
    
    
zipcode_info_dict['houses'] = num_of_houses_list
zipcode_info_dict['houses > 1000000'] = num_of_expensive_houses_list
zipcode_info_dict['proportion'] = percentage_of_expensive_houses
zipcode_info_dict['longitude'] = longitudes
zipcode_info_dict['latitude'] = latitudes
zipcode_df = pd.DataFrame(zipcode_info_dict, index = indices)
```
