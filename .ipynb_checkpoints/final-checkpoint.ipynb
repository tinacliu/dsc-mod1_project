### Step 2: Cleaning the data set

based on the above findings, we intend to do the following data cleaning

a) drop column view: as it contains mostly zeros which does not make intuitive sense

b) drop column sqrt_basement: as it contains invalid data '?'. Alternative choice here is that we turn the ? into Null and convert the data series into int. However given that this only contains part of the data about the size of the house and we have the more complete data seires sqrt_living available, we would not need to use sqrt_basement. For example, even if we cleaned sqrt_basement and it turns out to be highly correlated with house prices, it would be correlated with sqrt_living too, and we'd prefer to use sqrt_living in the model as it makes more intuitive sense as a better predictor.

c) convert date from string object to datetime

d) create a new binary data series based on yr_renovated, similar to waterfront. Intuitively it would make sense that recently renovated houses sell for more.

later on, before building the model, we will delete the rows that contain Nulls in waterfront.

```python
house_df['is_renovated'] = house_df['yr_renovated'].map(lambda x: 0 if not x else (0 if math.isnan(x) else 1))

clean_df = house_df.drop(columns = ['view','sqft_basement'])

clean_df.date = clean_df.date.map(lambda x: pd.to_datetime(x))
```

#### Question 1: Has houses got bigger over the years?

This question is likely best answered by plotting graphs, on the x-axis we will plot yr_built, on the y-axis we have several data series representing the size of the properties, we feel sqft_living, bedrooms are the most representitive of the house size.

We first group data by yr_built and plot the median sqft_living per each yr_built, this reduces/summerises the data which would enable a easier study on any trends present. We choose median rather than mean for sqft_living, as this is an continous data set with bounded min but somewhat unbounded max, and extreme large values could be present in the data which could skew the mean but not median. For bedrooms we chose to aggregate by mean as being a discrete dataset, the median is likely the same for many yr_built hence we would not be able to spot any trends.

We will also look at number of bathrooms as a supplement to the size trend.

```python

houses_by_year = clean_df.groupby('yr_built')
houses_in_year = houses_by_year.size()
```

```python
avg_live_size_by_year = clean_df.groupby('yr_built')['sqft_living'].median()
avg_room_size_by_year = clean_df.groupby('yr_built')['bedrooms'].mean()
avg_bath_size_by_year = clean_df.groupby('yr_built')['bathrooms'].mean()
```

```python
fig = plt.figure(figsize = (24,14))

ax1 = fig.add_subplot(221)
plt.scatter(avg_live_size_by_year.index, avg_live_size_by_year.values)
plt.title('Median Living Size by Built Year')

ax2 = fig.add_subplot(222)
plt.scatter(avg_room_size_by_year.index, avg_room_size_by_year.values)
plt.title('Average No. of Bedrooms by Built Year')

ax3 = fig.add_subplot(223)
plt.scatter(avg_bath_size_by_year.index, avg_bath_size_by_year.values)
plt.title('Average No. of Bathrooms by Built Year')

ax4 = fig.add_subplot(224)
plt.bar(houses_in_year.index, houses_in_year.values)
plt.title('Counts of property data per Built Year')

plt.show()
```

#### Question 1: Findings.

In looking at the number of data points we used in our aggregation (bottom right graph above), we see that data points is low during 1930 to 1940, which coincides with the decade of the great depression so it would make sense that not many houses was built during that period, and hence the count is low in our sales house price dataset (being a sample of the housing stock).

We keep this in mind when looking at the median size of properties (sqft_living), it looks like the size of properties built was stable from 1900 to 1940, with a slight dip for 1930 to 1940, however data points is also low here so we cannot be too confident about the dip. However it is clear that post 1950 (the second world war), houses built are increasing in an linear manner until more recently, post 2000 it looks like the size trend is on a slight decreasing slope. This is maybe not too surprising, given the more recent trend of younger generation wanting to live in city rather than surburbs, in flats rather than houses. However we do not have enough analysis nor data to conclude about the post-2000 trend.

Looking at the number of rooms, the trend is increasing from late 40s (post second world war) ending at mid 60s, this concides with the period of the Baby boomers, with an increasing birth rate observed in this period, families needed more bedrooms in houses. However post 60s, its unclear there is any trend. No. of bedrooms in houses built seem to be somewhat stable.

Number of bathrooms in houses built shows the strongest increasing trend post 40s to recently. With an average of 2.6 bathrooms for houses built post 2000, comparing to the 1.7 average bathroom level from 1900 to 1940, that is a 50% increase, and maybe suggests a lifestyle change for households in King County.

In summary, houses built after 1940s have been increasing in size as we thought in our hypothesis, but while the increase in average number of bedrooms in properties stopped post 1960s, the number of bathrooms continue to rise. Properties built post 1960s are getting bigger, with more bathrooms, and most likely bigger bedrooms.

#### Question 2: Findings.

Looking at the plots below, it seems that building condition has actually been declining over time. Checking the definition of building condition on King Country website(https://info.kingcounty.gov/assessor/esales/Glossary.aspx?type=r#b), building condition is the condition of the inside of the property, like wall surfaces, roofing, plumbing, heating, RELATIVE to age and grade. The work "relative" is key here as it means houses built in 1904 with mean of 3.8 is not directly comparable to the score of 3 given to houses build in 2014. 

Also, note by looking at the summary statistics above, we can see that older houses had a large standard deviation (around 1), meaning that while on average the mean is around 3.5, the range is from score of 1 or 2 to 5. Whereas for properties built more recently, all were given a score of 3.

Building grade records the quality of the construction, and as expected this does improve with time. The stardard deviation remains around the 1.0 level, suggesting the scoring system is at least somewhat consistent throught time.

We think the building grade data column is a better representation of the building quality, and this does indeed improve with time as we had assumed in the question/hypothesis.

```python
plt.figure(figsize=(16,14))

plt.subplot(2,2,1)
plt.title('Building Condition Throughtout Time')
plt.scatter(condition_yr.index, condition_yr[['mean']])


plt.subplot(2,2,2)
plt.title('Building Grade Throughtout Time')
plt.scatter(grade_yr.index, grade_yr[['mean']])

plt.subplot(2,2,3)
plt.title('Standard Dev. of Building Condition Score Throughtout Time')
plt.scatter(condition_yr.index, condition_yr[['std']])


plt.subplot(2,2,4)
plt.title('Standard Dev. of Building Grade Score Throughtout Time')
plt.scatter(grade_yr.index, grade_yr[['std']])
```

We next look to explore the relationship between the independent variables and dependent variable.

```python
# create a function to plot scatter plots between Y and independent variables
# note we plot graphs for all data columns but certain columns we dont plan to use as predictors 
# like id, date (not valid), lat, long, zipcode

def plot_data(dataframe, y):
    fig = plt.figure(figsize = (10,100))
    df_less_y = dataframe.drop(y, axis = 1)
    n = len(df_less_y.columns)
    cols = list(zip(range(n),df_less_y.columns))
    for i in range(n):
        ax = fig.add_subplot(round(n/2,0), 2, i+1)
        ax.scatter(df_less_y[cols[i][1]],dataframe[y])
        ax.set_title('Plot of {} against {}'.format(y,cols[i][1]))
        ax.set_xlabel(cols[i][1])
        ax.set_ylabel(y)
```

```python
plot_data(clean_df, 'price')
```

```python
# study the few houses that sold for more than 5 million

display(clean_df[clean_df['price'] >= 5_000_000])
# seems valid data points, with the number bedrooms, bathroom and area all makes sense
# hence no outliers in the Y variable

clean_df[(clean_df['price'] >= 3_000_000) & (clean_df['price'] < 5_000_000)].shape[0]
# another 44 houses sold for between 3 mil and 5 mil
# in total 51 houses sold for above 3 mil
```

From the plots, looks like sqft_living or sqft_living15, bathrooms and grade has the strongest relationship with price, where these x variables increase, y also increases. In the case of discret variables grade, as grade goes higher, the whole range of price points moves upwards.

However from above plots, its hard to say being on waterfront, and being renovated has any effect on the price.

Due to the amount of data points, its also hard to make out the trend, so we will try to aggregate the x variables.

```python

def plot_avgs_scatter(dataframe, group, variable ):
    grouped_means = dataframe.groupby(group).mean().reset_index()
    grouped_means.plot.scatter(group, variable)

def plot_med_scatter(dataframe, group, y_variable):
    grouped_means = dataframe.groupby(group).median().reset_index()
    grouped_means.plot.scatter(group, y_variable)
```

```python

plot_med_scatter(clean_df, 'bedrooms', 'price')
plt.ylim(0,2_000_000)
plt.show()

plt.bar(clean_df.bedrooms.value_counts().index, clean_df.bedrooms.value_counts().values)
plt.xlabel('no. of bedrooms')
plt.ylabel('counts')
plt.show()

# Note there are not enough data for the houses with 7 and higher bedrooms
# and relationshp up to 7 bedrooms is not very strong with price
```
