## Step 2: Cleaning the data set

based on the above findings, we intend to do the following data cleaning

a) drop column view: as it contains mostly zeros which does not make intuitive sense

b) drop column sqrt_basement: as it contains invalid data '?'. Alternative choice here is that we turn the ? into Null and convert the data series into int. For example, even if we cleaned sqrt_basement and it turns out to be highly correlated with house prices, it would be correlated with sqrt_living too, and we'd prefer to use sqrt_living in the model as it makes more intuitive sense as a better predictor.

#### Question 1: Has houses got bigger over the years?

This question is likely best answered by plotting graphs, on the x-axis we will plot yr_built, on the y-axis we have several data series representing the size of the properties, we feel sqft_living, bedrooms are the most representitive of the house size.

We first group data by yr_built and plot the median sqft_living per each yr_built, this reduces/summerises the data which would enable a easier study on any trends present. We choose median rather than mean for sqft_living, as this is an continous data set with bounded min but somewhat unbounded max, and extreme large values could be present in the data which could skew the mean but not median. For bedrooms we chose to aggregate by mean as being a discrete dataset, the median is likely the same for many yr_built hence we would not be able to spot any trends.

We will also look at number of bathrooms as a supplement to the size trend.

We next look to explore the relationship between the independent variables and dependent variable.

```python
# create a function to plot scatter plots between Y and independent variables
# note we plot graphs for all data columns but certain columns we dont plan to use as predictors 
# like id, date (not valid), lat, long, zipcode

def plot_data(dataframe, y):
    fig = plt.figure(figsize = (10,100))
    df_less_y = dataframe.drop(y, axis = 1)
    n = len(df_less_y.columns)
    cols = list(zip(range(n),df_less_y.columns))
    for i in range(n):
        ax = fig.add_subplot(round(n/2,0), 2, i+1)
        ax.scatter(df_less_y[cols[i][1]],dataframe[y], marker='.', alpha=0.1)
        ax.set_title('Plot of {} against {}'.format(y,cols[i][1]))
        ax.set_xlabel(cols[i][1])
        ax.set_ylabel(y)
```

```python
plot_data(clean_df.drop('date', axis = 1), 'price')
```

From the plots of x variables with price, looks like sqft_living or sqft_living15, bathrooms and grade has the strongest relationship with price, where these x variables increase, y also increases. ## Step 7: Further Insights into Zipcodes and Areas

Given that our model's relatively low prediction power (64% R2), it could only be used as a quide to a very rought price guideline. In order to provide more useful insights to business that are interested in the house price, for example house builder and real estate agents. 

One important factor that people care about when thinking about house purchase, is having to decide an area to live, chosen based on factors like affordability, 'feel' of the neighbourhood, ease of transport, schools etc. We intend to give more insights on the first two in the below analysis. Again, we are assuming our dataset is representive of the general housing stock existing.


```python

zipcode_info_dict = {}
med_prices = []
indices = []
num_of_expensive_houses_list = []
num_of_houses_list = []
longitudes = []
latitudes = []
percentage_of_expensive_houses = []


for code in clean_df.zipcode.unique():
    indices.append(code)
    med_prices.append(clean_df.groupby('zipcode').median().loc[code,'price'])
    num_of_expensive_houses = len(clean_df[(clean_df['zipcode'] == code) & (clean_df['price'] >= 1000000)])
    num_of_expensive_houses_list.append(num_of_expensive_houses)
    num_of_houses = len(clean_df[clean_df['zipcode'] == code])
    num_of_houses_list.append(num_of_houses)
    longitudes.append(
                    clean_df.groupby('zipcode').median().loc[code,'long'])
    latitudes.append(
                    clean_df.groupby('zipcode').median().loc[code,'lat'])
    percentage_of_expensive_houses.append(
                                        round((num_of_expensive_houses/num_of_houses*100),2))
    zipcode_info_dict['med_prices'] = med_prices
    
    
zipcode_info_dict['houses'] = num_of_houses_list
zipcode_info_dict['houses > 1000000'] = num_of_expensive_houses_list
zipcode_info_dict['proportion'] = percentage_of_expensive_houses
zipcode_info_dict['longitude'] = longitudes
zipcode_info_dict['latitude'] = latitudes
zipcode_df = pd.DataFrame(zipcode_info_dict, index = indices)
```
